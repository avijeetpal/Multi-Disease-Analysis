#### ðŸ“ŠData Analysis (Using Matplotlib)

In this step, I explored and analyzed the transformed dataset using **Matplotlib** to identify key patterns and insights.

#### ðŸŽ¯ Objectives:
- Identify which **age group** has more disease cases.
- Calculate **male vs female** percentage of cases.
- Find which **year** recorded the most disease cases.
- Analyze which **region** has the highest number of cases.
- Examine **disease type distribution** across years.

#### âœ… Analysis Performed:
- **Bar Chart:** Disease cases by age group.
- **Pie Chart:** Male vs Female disease ratio.
- **Bar Chart:** Year-wise case distribution.
- **Bar Chart:** Region-wise disease count.
- **Line Chart:** Yearly trend of each disease type.

#### ðŸ§  Insights Gained:
- Determined which **population group** (Young, Mid, Older, Senior) is most affected.
- Identified the **dominant gender** in disease occurrence.
- Found the **peak year** with maximum disease cases.
- Highlighted **regions** with the most reported cases.
- Observed the **trend of diseases** over time.

#### ðŸ“ˆ Outcome:
This analysis provided a complete understanding of how diseases vary  
by **age, gender, region, and year**, forming the base for Step 5 (Data Visualization in Power BI).


```python
# import required library

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt

# load tranformed file
multi = pd.read_csv('multi_disease_transformed.csv')
```

```python
# load basic Overview
print(multi.head())
print(multi.info())
print(multi.describe())
```

```python
# Age Group analysis

age_group_case = multi['Age Group'].value_counts()

plt.figure(figsize=(6,5))
plt.bar(age_group_case.index, age_group_case.values, color='blue')
plt.title('Disease Case by Age Group',fontweight='bold')
plt.xlabel('Age Group',fontweight='bold')
plt.ylabel('Number of Cases',fontweight='bold')
plt.grid(axis='y',linestyle = '--', alpha =0.3)

for i, value in enumerate(age_group_case.values):
    plt.text(i, value + 0.5, str(value), ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.show()
```

```python
# Gender Distribution (% Male vs Female)

gender_distribution = multi['Gender'].value_counts()
colors = ['coral', 'skyblue']
plt.figure(figsize=(5,5))
plt.pie(gender_distribution, labels=gender_distribution.index,autopct='%1.1f%%',startangle=90,colors=colors)
plt.title('Male vs Female Diseases Cases', fontweight='bold')
plt.show()
```

```python
# Yearly Disease Cases

year_case = multi['Year'].value_counts().sort_index()
plt.Figure(figsize=(6,6))
plt.bar(year_case.index,year_case.values, color='Orange')
plt.title('Yearly Disease Cases')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.grid(axis='y',linestyle='--',alpha=0.4)
for i, value in enumerate(year_case.values):
    plt.text(year_case.index[i], value + 0.5, str(value),
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.show()

```

```python
# region wise disease distribution

region_case = multi['Region'].value_counts()

plt.figure(figsize=(10,6))
plt.bar(region_case.index, region_case.values, color='mediumseagreen')
plt.title('Region wise disease distribution', fontweight='bold')
plt.xlabel('Region')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.grid(axis='y',linestyle='--',alpha=0.4)
for i, value in enumerate(region_case.values):
    plt.text(region_case.index[i], value + 0.5, str(value),
             ha='center', va='bottom', fontsize=10, fontweight='bold')
plt.show()
```

```python
# disease trend per year

plt.figure(figsize=(10,6))
for disease in multi['Disease'].unique():
    subset = multi[multi['Disease'] == disease]
    yearly_trend = subset['Year'].value_counts().sort_index()
    plt.plot(yearly_trend.index, yearly_trend.values, marker='o', label=disease)

plt.title('Disease Trend per Year')
plt.xlabel('Year')
plt.ylabel('Number of Cases')
plt.legend(title='Disease Type')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
# Summary Insights

print("Most Affected Age Group:", multi['Age Group'].value_counts().idxmax())
print("Most Affected Gender:", multi['Gender'].value_counts().idxmax())
print("Year with Maximum Cases:", multi['Year'].value_counts().idxmax())
print("Region with Maximum Cases:", multi['Region'].value_counts().idxmax())
print("Most Common Disease:", multi['Disease'].value_counts().idxmax())
```
